```python
import keyboard
import cv2
```

```python


face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml') 

eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml') 

palm_cascade = cv2.CascadeClassifier('closed_palm.xml')


cap = cv2.VideoCapture(0) 
play = True 
# to count 120 frames continously 
counter = 0
skip = False
# face flag to keep track the space has been already pressed
no_face = False
#to store the face position so we can relativly differentiate between  the right palm or left palm
x_pos_face = 0

while 1: 

    # reads frames from a camera 
    ret, img = cap.read() 
    
    # convert to gray scale of each frames 
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

    # Detects faces of different sizes in the input image 
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) 
    palm = palm_cascade.detectMultiScale(gray, 1.3, 5)
    
    
    # cheking the codition of face present or not 
    if(len(faces) == 0 and no_face == False ):
        keyboard.press_and_release('space')
        no_face = True
        #skip = True
    elif(len(faces)  != 0 and no_face == True):
        keyboard.press_and_release('space')
        no_face = False
        #skip = False

    for (x,y,w,h) in faces: 
        
        x_pos_face = x
        
        # To draw a rectangle in a face 
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2)  
        roi_gray = gray[y:y+h, x:x+w] 
        roi_color = img[y:y+h, x:x+w] 
        
        

        # Detects eyes of different sizes in the input image 
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if(len(eyes) == 0):
            counter = counter + 1
        else:
            counter = 0
            
        if(play == True and counter > 120):
            keyboard.press_and_release('space')
            play = False
            if(counter == 10000):
                counter = 121
        
        if(play == False and counter == 0):
            keyboard.press_and_release('space')
            play = True
        
        

        #To draw a rectangle in eyes 
        for (ex,ey,ew,eh) in eyes: 
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,127,255),2) 

            
    for (x,y,w,h) in palm:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2)  
        roi_gray = gray[y:y+h, x:x+w] 
        roi_color = img[y:y+h, x:x+w] 
        
        if(x<x_pos_face):
            keyboard.press_and_release('Ctrl+Up')
        else:
            keyboard.press_and_release('Ctrl+Down')
        
    # Display an image in a window 
    cv2.imshow('img',img) 

    # Wait for Esc key to stop 
    k = cv2.waitKey(30) & 0xff
    if k == 27: 
        break

# Close the window 
cap.release() 

# De-allocate any associated memory usage 
cv2.destroyAllWindows() 
```
